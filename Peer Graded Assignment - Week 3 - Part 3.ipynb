## Lets cluster the data based on the Boroughs and Neighbourhoods.

### We need to look at the number of boroughs and neighbourhoods

```python
print('Toronoto has {} boroughs and {} neighborhoods.'.format(len(df['Borough'].unique()),df.shape[0]))
```

### Let us cluster these burrows based on their vicinity 
#### I will choose k = 6 as it will evenly space out the Toronto Map 


```python
#The co-ordinates will tell us about the vicinity of the place
X = df['Latitude']
Y = df['Longitude']
Z = np.stack((X, Y), axis=1)

#Fit Z into the K-Means Algorithm to get our clusters
kmeans = KMeans(n_clusters=6, random_state=0).fit(Z)
clusters = kmeans.labels_
clusters
```

#### Now that we have our clusters in place,

# LET'S VISUALIZE!

```python
toronto_map = folium.Map(location=[43.65, -79.4], zoom_start=12)

clusters = kmeans.labels_
colors = ['red', 'green', 'blue', 'yellow', 'cyan', 'pink']
df['Cluster'] = clusters

for latitude, longitude, borough, cluster in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Cluster']):
    label = folium.Popup(borough, parse_html=True)
    folium.CircleMarker(
        [latitude, longitude],
        radius=5,
        popup=label,
        color='black',
        fill=True,
        fill_color=colors[cluster],
        fill_opacity=0.7).add_to(toronto_map)  

toronto_map
```
